## Exemplo em um Grafo Simples

```python
import networkx as nx
import numpy as np
```

```python
from collections import deque
```

```python
G = nx.Graph()
```

```python
G.add_edges_from(
    [('v','r'),
     ('r','s'),
     ('s','w'),
     ('w','t'),
     ('w','x'),
     ('t','x'),
     ('t','u'),
     ('x','u'),
     ('x','z'),
     ('u','z')
     ]
)
```

```python
nx.draw(G, with_labels=True)
```

```python
s = 's'
```

```python
G.nodes()
```

```python
G.edges()
```

```python
# def BFS(G,s):

# INICIALIZACAO
for v in G.nodes() - {s}:
    G.nodes[v]['cor'] = 'branco'
    G.nodes[v]['dis'] = np.inf

G.nodes[s]['cor'] = 'cinza'
G.nodes[s]['dis'] = 0

# Fila (append (right), popleft)
Q = deque()
Q.append(s)
```

```python
G.nodes(data=True)
```
